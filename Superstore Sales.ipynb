### 2. Create a function called remove_outliers which takes as parameter the dataframe and the column which needs to be searched for outliers.

```python
## Creating a function called remove_outliers which takes as parameter the dataframe and the column which needs to be searched for outliers:
def remove_outliers (dataframe_name, column_name):
    Q1 = dataframe_name[column_name].quantile(0.25)
    Q2 = dataframe_name[column_name].quantile(0.50)
    Q3 = dataframe_name[column_name].quantile(0.75)
    IQR = Q3-Q1
    upper_bound = Q3 + 3*IQR
    lower_bound = Q1 - 3*IQR
    df = dataframe_name[(dataframe_name[column_name]>upper_bound) | (dataframe_name[column_name]<lower_bound)]
    if (df.shape[0]/dataframe_name.shape[0])*100 > 2:
        if (dataframe_name[(dataframe_name[column_name]>Q1) & (dataframe_name[column_name]<Q2)].shape[0]) == (dataframe_name[(dataframe_name[column_name]>Q2) & (dataframe_name[column_name]<Q3)].shape[0]):
            for i in dataframe_name.index:  
                if (dataframe_name.loc[i,column_name]>upper_bound) | (dataframe_name.loc[i,column_name]<lower_bound):
                    dataframe_name.loc[i,column_name]=dataframe_name[column_name].mean()
        else:
            for i in dataframe_name.index:  
                if (dataframe_name.loc[i,column_name]>upper_bound) | (dataframe_name.loc[i,column_name]<lower_bound):
                    dataframe_name.loc[i,column_name]=dataframe_name[column_name].median()
    else:
         for j in dataframe_name.index:  
                if (dataframe_name.loc[j,column_name]>upper_bound) | (dataframe_name.loc[j,column_name]<lower_bound):
                    dataframe_name.drop(index=j,inplace = True)
```

```python
## Removing outliers in the Sales Price column using remove_outliers function:
remove_outliers(Merge_data, 'Sales Price')
```

### 1. Sales and Profit Trends Over Time: Use a time series plot to analyze how sales and profit have trended over the years or months.

```python
## Creating a new column containing both the year and the month:
data['year'] = data['Order Date'].dt.year
data['month'] = data['Order Date'].dt.month
data['yr-month'] = data['year'].astype(str)+'-'+data['month'].astype(str)
```

```python
## Displaying the data:
data
```

```python
## Creating a data containing total sales and total profit year-wise:
yr_data = data.groupby('year').aggregate({'Total Sales':'sum','Total Profit':'sum'})
## Restting the index:
yr_data.reset_index(inplace = True)

## Creating a data containing total sales and total profit month-wise:
month_data = data.groupby('yr-month').aggregate({'Total Sales':'sum','Total Profit':'sum'})
## Restting the index:
month_data.reset_index(inplace = True)
```

```python
## Plotting a line chart to display the trends of Total Sales and Total Profit over the years:
yr_data.plot(x='year',y=['Total Sales','Total Profit'],marker = 'v',color=['brown','green'],figsize=(10,6))
plt.title('Trends of Total Sales and Total Profit over the years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.grid(True)
plt.show()
```

```python
## Plotting a line chart to display the trends of Total Sales and Total Profit over the months:
month_data.plot(x='yr-month',y=['Total Sales','Total Profit'],marker = 'v',color=['brown','green'],figsize=(10,6))
plt.title('Trends of Total Sales and Total Profit over the months')
plt.xlabel('Year-Month')
plt.ylabel('Amount')
plt.grid(True)
plt.show()
```

### 3. Yearly Growth in Sales and Profit: Use a year-over-year growth chart to compare the sales and profit growth over different years.

```python
## Creating the data containing total sales and total profit over the years:
yr_data = data.groupby('year').aggregate({'Total Sales':'sum','Total Profit':'sum'})

## Resetting the index:
yr_data.reset_index(inplace = True)
```

```python
## Displaying the data:
yr_data
```

```python
## Creating a new column containing the growth in sales:
for i in range(yr_data.index[0]+1,yr_data.index[-1]+1):
    yr_data.loc[i,'Growth in Sales'] = yr_data.loc[i,'Total Sales']-yr_data.loc[i-1,'Total Sales']
```

```python
## Creating a new column containing the growth in profit:
for i in range(yr_data.index[0]+1,yr_data.index[-1]+1):
    yr_data.loc[i,'Growth in Profit'] = yr_data.loc[i,'Total Profit']-yr_data.loc[i-1,'Total Profit']
```

```python
## Displaying the data:
yr_data
```

```python
## Creating the data containing the growth in sales and profit:
growth_data = yr_data[['year','Growth in Sales','Growth in Profit']]
growth_data['Growth in Sales'] = growth_data['Growth in Sales'].fillna(0)
growth_data['Growth in Profit'] = growth_data['Growth in Profit'].fillna(0)

## Displaying the data:
growth_data
```

```python
## Plotting a line chart to show the growth in sales and profit:
growth_data.plot(x='year',y=['Growth in Sales','Growth in Profit'],marker = 'v',color=['brown','green'],figsize=(10,6))
plt.title('Growth in Sales and Profit over years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.grid(True)
plt.show()
```
